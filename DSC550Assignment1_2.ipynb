```python
#import libraries
import numpy as np
import pandas as pd
```

```python
#Load dataframe from CSV file
df = pd.read_csv("Video_Games_Sales_as_at_22_Dec_2016.csv")
```

```python
# Display first 10 rows of data
df.head(10)
```

```python
#calculating row and column numbers, there are 16719 rows and 16 columns
df.shape
```

```python
#Sort values by Critic score and show head of 5 top rated

df.sort_values(by=["Critic_Score"], ascending = False).head(5)
```

```python
#Calculating Genre Counts
genre_df = df[['Name','Genre']]
genre_df.groupby(["Genre"]).count()
```

```python
#First five games released on SNES
SNES_df = df.loc[df['Platform'] == 'SNES']
SNES_df.sort_values(by = ["Year_of_Release"], ascending = True).head(5)
```

```python
#calculating global sales by Publisher
global_sales = df[['Global_Sales',"Publisher"]]
global_sales.groupby(["Publisher"]).sum().sort_values(by = ["Global_Sales"], ascending = False).head(5)
```

```python
#adding new column that calculates North American Sales as a percentage of Global Sales
df['NA_%_Of_Global_Sales'] = (df['NA_Sales']/df['Global_Sales'] * 100)
df.head(5)
```

```python
#sum of NaNs in each column
df.isna().sum()
```

```python
#calculating median to find string error
df["User_Score"].median()
```

```python
#replacing string and calculating median
df["User_Score"] = df['User_Score'].replace('tbd', np.nan)
median = df['User_Score'].median()
print("Median User Score:", median)
```

```python
#replacing NaNs with median
df["User_Score"].fillna(value = median, inplace=True)
df.head(20)
```
